Question 1: What is hypothesis testing in statistics?
- Hypothesis testing is a fundamental concept in statistics used to make decisions or inferences about a population based on a sample of data. It involves testing an assumption (called a hypothesis) about a population parameter.
- Hypothesis testing is a statistical method used to make decisions about a population parameter based on sample data. It involves formulating two opposing hypotheses — the null hypothesis (H₀) and the alternative hypothesis (H₁) — and using sample data to determine which hypothesis is more supported by the evidence.
- Steps in Hypothesis Testing:
  - State the hypotheses (H₀ and H₁).
  - Choose a significance level (α):
  - Common values: 0.05, 0.01
  - This is the probability of rejecting H₀ when it is actually true (Type I error).
  - Collect and analyze sample data.
  - Calculate a test statistic:
  - Depends on the test used (e.g., z-test, t-test, chi-square test).
  - Determine the p-value:The probability of observing the data (or something more extreme) if H₀ is true.
  - Compare the p-value with α:
    - If p ≤ α: Reject H₀ (evidence supports H₁).
    - If p > α: Fail to reject H₀ (not enough evidence to support H₁).

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
- 1)**Null Hypothesis (H₀)**: This is the default or original assumption.It typically states that there is no effect or no difference.
- The null hypothesis is a statement that there is no effect, no difference, or no relationship in the population. It represents the default or original assumption that we aim to test.
- Example: 1)The average height of adult men in a city is 175 cm. 2)The average test score is 70. 3)It is assumed to be true unless there is strong evidence against it.
- 2)**Alternative Hypothesis (H₁ or Ha)**: This is what you want to test or prove.It represents a new claim or an effect/difference.
- The alternative hypothesis is a statement that contradicts the null hypothesis. It suggests that there is an effect, a difference, or a relationship.
- Example: 1)The average height of adult men in the city is not 175 cm.2)The average test score is not 70. 3)It is what the researcher typically wants to prove.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- The **significance level**, often denoted by α (alpha), is a threshold used in hypothesis testing to determine whether to reject the null hypothesis (H₀).
- The significance level (α) is the probability of making a Type I error, which means rejecting the null hypothesis when it is actually true.
- Common values: 1)α = 0.05 (5%) 2)α = 0.01 (1%)  3)α = 0.10 (10%)
- Role in Hypothesis Testing:
  - Set before the test begins, α defines how much risk you're willing to take of incorrectly rejecting H₀.
  - After conducting the test, you calculate a p-value — the probability of observing your data (or more extreme) if H₀ is true.
- Decision Rule:
  - If p-value ≤ α → Reject H₀ (evidence is statistically significant)
  - If p-value > α → Fail to reject H₀ (evidence is not strong enough)
- Example: Suppose you're testing a new medicine.
   - H₀: The medicine has no effect.
   - H₁: The medicine has an effect.
   - α = 0.05
   - After testing, you get a p-value = 0.03.
   - Since 0.03 < 0.05, you reject H₀, suggesting the medicine does have an effect.

- The significance level helps control false positives. A lower α (like 0.01) means you're being more strict and want stronger evidence before rejecting H₀.

Question 4: What are Type I and Type II errors? Give examples of each.
- When making decisions in hypothesis testing, two kinds of errors can occur:
- 1. **Type I Error (False Positive):** Rejecting the null hypothesis (H₀) when it is actually true. It means you concluded there is an effect or difference when there really isn't.Probability of Type I Error = α (significance level).
- Example of Type I Error: You test a new drug to see if it improves patient recovery.
  - H₀: The drug has no effect.

  - H₁: The drug has an effect.

  - You reject H₀ and claim the drug works, but in reality, it doesn’t.

  - Type I Error: You approved a drug that doesn’t work — a false positive.
- 2. **Type II Error (False Negative)**:Failing to reject the null hypothesis (H₀) when it is actually false.It means you concluded there is no effect, but there actually is one. Probability of Type II Error = β.
- Example of Type II Error: You test the same drug again.
  - H₀: The drug has no effect.

  - H₁: The drug has an effect.

  - This time, you fail to reject H₀, so you say the drug doesn’t work — but it actually does.
  -  Type II Error: You missed the opportunity to use an effective drug — a false negative.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
- The Z-test and T-test are both statistical methods used to determine whether there is a significant difference between groups or to test hypotheses about population parameters. However, they are used in different situations, primarily depending on sample size and population standard deviation knowledge.
- **Z-test**
  - Use when: The population standard deviation (σ) is known.The sample size is large (typically n ≥ 30).The data is approximately normally distributed (especially important for small samples).
  - Common applications: Testing the mean of a population when σ is known. Comparing proportions between two large samples.
  - Example: You're testing whether the average height of adult men in a city differs from the national average, and you know the population standard deviation.
- **T-test**
  - Use when: The population standard deviation is unknown.The sample size is small (typically n < 30).The data is approximately normally distributed.
  - Types of T-tests:One-sample t-test – Compare sample mean to a known value.Two-sample t-test – Compare means of two independent samples.Paired sample t-test – Compare means from the same group at different times (before/after).
  - Example:You're comparing the average test scores of a class of 15 students to a national benchmark, but you don’t know the population standard deviation.
- Use a Z-test when the sample is large and you know the population standard deviation.
- Use a T-test when the sample is small or the population standard deviation is unknown.

```python
"""
Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
"""
import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


population_mean = 50

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Significance level
alpha = 0.05

# Output
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")

```

```python
"""
Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
"""

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate data from a normal distribution
np.random.seed(42)
mean = 100
std_dev = 15
sample_size = 100

data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Step 2: Calculate the 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
confidence_level = 0.95
alpha = 1 - confidence_level

# t-distribution critical value
t_crit = stats.t.ppf(1 - alpha/2, df=sample_size - 1)

margin_of_error = t_crit * (sample_std / np.sqrt(sample_size))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 3: Plot the data and confidence interval
plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='--', label=f'95% CI Lower = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'95% CI Upper = {ci_upper:.2f}')
plt.title('Histogram of Simulated Normal Data with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
"""
Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
"""

import numpy as np
import matplotlib.pyplot as plt

def z_score_standardization(data):
    """
    Calculate Z-scores for a dataset and plot the standardized data as a histogram.

    Parameters:
    data (list or numpy array): The input dataset.

    Returns:
    z_scores (numpy array): The standardized Z-scores of the dataset.
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # histogram
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=20, color='skyblue', edgecolor='black')
    plt.title("Histogram of Z-Scores (Standardized Data)")
    plt.xlabel("Z-Score")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=1.5, label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores


# Example
data = [100, 102, 98, 105, 110, 95, 90, 115, 120, 85, 130]

z_scores = z_score_standardization(data)
print("Z-Scores:", z_scores)
```
